### 페이지 넘기기

```python
start = time.time()

title = []
link = []
date = []


path = '../driver/chromedriver.exe'
driver = webdriver.Chrome(path)

for page in range(1,81):
    url='https://search.daum.net/search?w=news&enc=utf8&cluster=y&cluster_page=1&q=%EB%94%94%EC%98%AC&cpname=%EC%97%98%EB%A5%B4&cp=16Y6Oklp6xw4QZhCK0&DA=PGD&p='+str(page)
    driver.get(url)
    
    print(driver.current_url)
    
    elements = driver.find_elements_by_css_selector('#clusterResultUL > li')
    
    for element in elements:
        title.append(element.find_element_by_css_selector('div.wrap_cont > div > div > a').text)

        link.append(element.find_element_by_css_selector('div.wrap_cont > div > span.f_nb.date > a').get_attribute('href'))

        date.append(element.find_element_by_css_selector('div.wrap_cont > div > span.f_nb.date').text[:10])
        
        
print(time.time() - start)
```

```python
elle_dior = pd.DataFrame({
    'title':title,
    'link':link,
    'date':date
})

elle_dior.head()
```

```python
elle_dior['date'].unique()
```

```python
elle_dior['date'] = elle_dior['date'].str.replace('.', '')
```

```python
elle_dior['date'] = elle_dior['date'].astype(int)
```

```python
elle_dior.loc[elle_dior['date']<20160901, 'date'] = np.NaN
```

```python
elle_dior.isnull().sum()
```

```python
elle_dior.dropna()
```

```python
elle_dior.to_csv('elle_dior.csv', sep = '|', encoding = 'utf-8', index = False)
```
