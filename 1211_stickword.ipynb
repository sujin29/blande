```python
# 걸크러시
mag_jenny['ADFN'] = mag_jenny['ADFN'].replace("'걸', '크러시", "걸크러시")
# 걸그룹
mag_jenny['ADFN'] = mag_jenny['ADFN'].replace("'걸', '그룹'", "걸그룹")
```

```python
# 글로우
mag_jenny['ADFN'] = mag_jenny['ADFN'].replace("'글', '로우'", "글로우")
```

```python
# 아우터
mag_jenny['ADFN'] = mag_jenny['ADFN'].replace("'아우', '터'", "아우터")
```

```python
# 데일리
mag_jenny['ADFN'] = mag_jenny['ADFN'].replace("'데', '일리'", "데일리")
```

```python
# 볼륨감
mag_jenny['ADFN'] = mag_jenny['ADFN'].replace("'볼륨', '감'", "볼륨감")
# 존재감
mag_jenny['ADFN'] = mag_jenny['ADFN'].replace("'존재', '감'", "존재감")
# 통일감
mag_jenny['ADFN'] = mag_jenny['ADFN'].replace("'통일', '감'", "통일감")
```

```python
# 스타일링
mag_jenny['ADFN'] = mag_jenny['ADFN'].str.replace("'스타', '일링'", "스타일링")
```

```python
# 캐주얼
mag_jenny['ADFN'] = mag_jenny['ADFN'].str.replace("'캐, '주얼'", "캐주얼")
```

```python
# FW / SS
mag_jenny['ADFN'] = mag_jenny['ADFN'].str.replace("'F', 'W'", "FW")
mag_jenny['ADFN'] = mag_jenny['ADFN'].str.replace("'S', 'S'", "SS")
```

```python
# 스포티
mag_jenny['ADFN'] = mag_jenny['ADFN'].str.replace("'스포', '티'", "스포티")
```

```python
# 테일러 / 레이어
mag_jenny['ADFN'] = mag_jenny['ADFN'].str.replace("'테일러', '드'", "테일러드")
mag_jenny['ADFN'] = mag_jenny['ADFN'].str.replace("'레이어', '드'", "레이어드")
```
