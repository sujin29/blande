### 함수 적용

```python
magazines_ce = [a_ce, b_ce, c_ce, e_ce, fb_ce, fn_ce, m_ce, w_ce]
```

```python
def sort_by_date(df):
    df.sort_values(by='date', ascending = False, inplace = True)
    df.reset_index(drop=True, inplace = True)
```

```python
for mag in magazines_ce:
    sort_by_date(mag)
```

## check 2 제외

```python
def drop_2(df):
    return df[df['check'] != 2]
```

```python
a_ce_d = drop_2(a_ce)
b_ce_d = drop_2(b_ce)
c_ce_d = drop_2(c_ce)
e_ce_d = drop_2(e_ce)
fb_ce_d = drop_2(fb_ce)
fn_ce_d = drop_2(fn_ce)
m_ce_d = drop_2(m_ce)
w_ce_d = drop_2(w_ce)
```

## list(article_list) 생성 후 append

```python
contents_list = []
```

```python
magazines_ce_d = [a_ce_d, b_ce_d, c_ce_d, e_ce_d, fb_ce_d, fn_ce_d, m_ce_d, w_ce_d]
```

```python
magazine_list = []
article_list = []
for magazine in magazines_ce_d:
    articles = magazine['contents']
    for article in articles:
        article_list.append(article)
```

## 트위터 패키지 안에 konlpy 모듈 호출

```python
from konlpy.tag import Twitter
```

```python
twitter = Twitter()
```

### 각 문장별로 형태소 구분하기

```python
sentences_tag = []

for sentence in article_list:
    morph = twitter.pos(sentence)
    sentences_tag.append(morph)
    
    

print(len(sentences_tag))
```
