```python
CH8R['ADFN_2'] = CH8R['ADFN_2'].str.replace('[', '').str.replace(']', '').str.replace('"', '').str.replace("'", "").str.split(', ')
```

```python
noun_adj_list = []
for i in range(len(CH8R)):
    for word in CH8R.loc[i, 'ADFN_2']:
        if word not in stop_words:
            noun_adj_list.append(word)
```

```python
word = Counter(noun_adj_list)
print(word.most_common(200))
```

```python
count = Counter(noun_adj_list)
```

```python
word_list = count.most_common(500)
for v in word_list:
    print(v)
```

```python
top500 = pd.DataFrame(word_list)
```

```python
top500.columns = ['word', 'count']
```
