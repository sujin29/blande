```python
link_list = []

path = '../driver/chromedriver.exe'
driver = webdriver.Chrome(path)
url = 'https://news.hmgjournal.com/Search-Result/List?newsearchKey=%EC%A0%9C%EB%84%A4%EC%8B%9C%EC%8A%A4&searchKey=&searchMenuId=AL'
driver.get(url)

video = driver.find_element_by_css_selector('#contents > ul > li:nth-child(4) > a')
video.click()

nextBtn = driver.find_element_by_css_selector('#contents > article > div > div > a')
isFlag = nextBtn.is_enabled()
print(isFlag)

while isFlag == True:
    if nextBtn.text == '더보기':
        print('next page exist')
        nextBtn.click()
        
        time.sleep(1)
    else:
        break
        
contElements = driver.find_elements_by_css_selector('#listImg > li')

for cont in contElements:
#     # 타이틀 가져오기
#     articles = cont.find_element_by_css_selector('#listImg > li.firstContent > a > span').get_attribute('title')
#     print(articles)
    
    # 링크 가져오기
    link = cont.find_element_by_tag_name('a').get_attribute('href')
    link_list.append(link)
    
    
len(link_list)
```

```python
title = []
tag = []
contents = []

path = '../driver/chromedriver.exe'
driver = webdriver.Chrome(path)

for link in hmg_video['link']:
    url = link
    driver.get(url)
    
    print(driver.current_url)
    
    # 타이틀 가져오기
    title.append(driver.find_element_by_css_selector('#contents > div.area_video_txt > div.box_video_txt > h2').text)
    
    # 태그 가져오기
    tag.append(driver.find_element_by_tag_name('#contents > div.area_video_txt > div.box_video_txt > p').text)
    
    # 본문 가져오기
    contents.append(driver.find_element_by_tag_name('#text').text)
    
    time.sleep(1)
```

```python
print('title', len(title))
print('tag', len(tag))
print('contents', len(contents))
```

```python
hmg_video['title'] = title
hmg_video['tag'] = tag
hmg_video['contents'] = contents
```

```python
hmg_video = hmg_video[['title', 'link', 'tag', 'contents']]
```
