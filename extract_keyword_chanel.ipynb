```python
len(a_ch_d)
```

```python
a_ch_list = []
for i in range(len(a_ch_d)):
    cont = []
    if a_ch_d.loc[i, 'title_brand'] == 1:
        cont.append(a_ch_d.loc[i, 'contents'])
    else:
        for content in a_ch_d.loc[i, 'contents'].split("', '"):
#             print(content)
#             print('-'*50)
            if '샤넬' in content:
                cont.append(content)
    a_ch_list.append(cont)

len(a_ch_list)
```

```python
a_ch_d['word_celine'] = a_ch_list
```
