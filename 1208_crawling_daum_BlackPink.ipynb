## 엘르

```python
start = time.time()

title = []
link = []
date = []
elle_link = []


path = '../driver/chromedriver.exe'
driver = webdriver.Chrome(path)

for page in range(1,13):
    url='https://search.daum.net/search?nil_suggest=btn&w=news&cluster=y&q=%EC%A0%9C%EB%8B%88&sd=20160701000000&ed=20201208235959&period=u&cpname=%EC%97%98%EB%A5%B4&cp=16Y6Oklp6xw4QZhCK0&DA=PGD&p='+str(page)
    driver.get(url)
    
    print(driver.current_url)
    
    elements = driver.find_elements_by_css_selector('#clusterResultUL > li')
    
    for element in elements:
        title.append(element.find_element_by_css_selector('div.wrap_cont > div > div > a').text)

        link.append(element.find_element_by_css_selector('div.wrap_cont > div > span.f_nb.date > a').get_attribute('href'))

        date.append(element.find_element_by_css_selector('div.wrap_cont > div > span.f_nb.date').text[:10])
        
        elle_link.append(element.find_element_by_css_selector('div.wrap_cont > div > div > a').get_attribute('href'))
        
        
print(time.time() - start)
```

```python
elle_jenny = pd.DataFrame({
    'title':title,
    'link':link,
    'date':date,
    'elle_link':elle_link
})

elle_jenny.head()
```

## 샤넬

```python
start = time.time()

content_list = []

path = '../driver/chromedriver.exe'
driver = webdriver.Chrome(path)

for i in range(len(elle_jenny)) :
    url = elle_jenny['link'][i]
    driver.get(url)
    
    print(i, driver.current_url)
    
    div_list = []
    p_list = []
    
    contents = driver.find_elements_by_css_selector('#harmonyContainer > section')

    for content in contents:
        try:
            divs = content.find_elements_by_css_selector('div')
            for div in divs:
                div_text = div.text
                div_list.append(div_text)
        
        except NoSuchElementException as nse :
            pass
        
        try:
            ps = content.find_elements_by_css_selector('p')
            for p in ps:
                p_text = p.text
                p_list.append(p_text)
            
        except NoSuchElementException as nse :
            pass
            
            

        

        
    content_list.append([div_list, p_list])
    
    
    time.sleep(10)

    


    

        
print('-'*50)
print(len(content_list))
print(time.time() - start)



#harmonyContainer > section > p
#harmonyContainer > section > div
#harmonyContainer > section > div
```
