```python
before = mag_ce[mag_ce['date'].str[:4].astype(int)<2019]
after = mag_ce[mag_ce['date'].str[:4].astype(int) >= 2019]

before.reset_index(drop = True, inplace = True)
after.reset_index(drop = True, inplace = True)
```

```python
ADFN_word_list = []
for i in range(len(mag_ce)):
    for word in mag_ce.loc[i, 'ADFN']:
        if word not in stop_words:
            ADFN_word_list.append(word)
```

```python
count = Counter(ADFN_word_list)
print(count.most_common(100))
```

```python
common = [ '모델', '백', '브랜드',  '쇼' ,'파리', '디자인', '디자이너', 'COUTURE', '세계', '장식', '드레스', '그녀',
          '여성', '봄', '여름', '가을', '겨울', '광고', '시절', '패션', '새로운', '사진'] 
```

```python
before_word_list = []
for i in range(len(before)):
    for word in before.loc[i, 'ADFN']:
        if word not in stop_words:
            if word not in common:
                before_word_list.append(word)
```

```python
b_count = Counter(before_word_list)
print(b_count.most_common(100))
```

```python
after_word_list = []
for i in range(len(after)):
    for word in after.loc[i, 'ADFN']:
        if word not in stop_words:
            if word not in common:
                after_word_list.append(word)
```

```python
a_count = Counter(after_word_list)
print(a_count.most_common(100))
```

```python
b_word_list = b_count.most_common(500)
a_word_list = a_count.most_common(500)
```
