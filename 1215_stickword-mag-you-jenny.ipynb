```python
tot_je['ADFN_2'] = tot_je['ADFN_2'].str.replace('[', '').str.replace(']', '').str.replace('"', '').str.replace("'", "").str.split(', ')
```

```python
noun_adj_list = []
for i in range(len(tot_je)):
    for word in tot_je.loc[i, 'ADFN_2']:
        if word not in stop_words:
            noun_adj_list.append(word)
```
