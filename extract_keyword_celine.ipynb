```python
a_ce_list = []
for i in range(len(a_ce_d)):
    cont = []
    if a_ce_d.loc[i, 'title_brand'] == 1:
        cont.append(a_ce_d.loc[i, 'contents'])
    else:
        for content in a_ce_d.loc[i, 'contents'].split("', '"):
#             print(content)
#             print('-'*50)
            if ('셀린' or '셀린느') in content:
                cont.append(content)
    a_ce_list.append(cont)

len(a_ce_list)
```

```python
a_ce_d['word_celine'] = a_ce_list
```
