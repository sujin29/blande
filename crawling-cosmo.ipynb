### 셀린느

```python
start = time.time()

title = []
cosmo_link = []
date = []

path = '../driver/chromedriver.exe'
driver = webdriver.Chrome(path)

for page in range(1,13):
    url='https://search.daum.net/search?nil_suggest=btn&w=news&cluster=y&q=%EC%85%80%EB%A6%B0%EB%8A%90&sd=20160901000000&ed=20201123235959&period=u&cpname=%EC%BD%94%EC%8A%A4%EB%AA%A8%ED%8F%B4%EB%A6%AC%ED%83%84&cp=16UGJQPoOyBSIgo3a1&DA=PGD&p='+str(page)
    driver.get(url)
    
    print(driver.current_url)
    
    elements = driver.find_elements_by_css_selector('#clusterResultUL > li')
    
    for element in elements:
        title.append(element.find_element_by_css_selector('div.wrap_cont > div > div > a').text)

        cosmo_link.append(element.find_element_by_css_selector('div.wrap_cont > div > div > a').get_attribute('href'))
        
        date.append(element.find_element_by_css_selector('div.wrap_cont > div > span.f_nb.date').text[:10])

        
print('-'*50)        
print(time.time() - start)

cosmo_link_chanel = pd.DataFrame({
    'title': title,
    'cosmo_link':cosmo_link,
    'date':date
})
```

```python
cosmo_link_celine = cosmo_link_chanel
```

```python
len(cosmo_link_celine)
```

```python
cosmo_link_celine.head()
```

```python
cosmo_link_celine['date'] = cosmo_link_celine['date'].str.replace('.', '-')
```

```python
cosmo_celine = pd.merge(cosmo_celine, cosmo_link_celine, on = ['title', 'date'], how='outer')
```

```python
cosmo_celine.head()
```

```python
path = '../driver/chromedriver.exe'
driver = webdriver.Chrome(path)

category = []


for i in range(len(cosmo_celine)) :
    url = cosmo_celine['cosmo_link'][i]
    driver.get(url)
    
    print(i, driver.current_url)
    
    try:
        category.append(driver.find_element_by_css_selector('#contents > div > div.article_head > div').find_element_by_tag_name('span').text)
        
    except NoSuchElementException as nse :
        category.append(0)
    
    time.sleep(2)
    
    
print('-'*50)
print(len(category))
```
