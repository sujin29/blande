### 함수 적용

```python
magazines_ch = [a_ch, b_ch, c_ch, e_ch, fb_ch, fn_ch, m_ch, v_ch, w_ch]
```

```python
def sort_by_date(df):
    df.sort_values(by='date', ascending = False, inplace = True)
    df.reset_index(drop=True, inplace = True)
```

```python
for mag in magazines_ch:
    sort_by_date(mag)
```

## check 2 제외

```python
def drop_2(df):
    return df[df['check'] != 2]
```

```python
a_ch_d = drop_2(a_ch)
b_ch_d = drop_2(b_ch)
c_ch_d = drop_2(c_ch)
e_ch_d = drop_2(e_ch)
fb_ch_d = drop_2(fb_ch)
fn_ch_d = drop_2(fn_ch)
m_ch_d = drop_2(m_ch)
v_ch_d = drop_2(v_ch)
w_ch_d = drop_2(w_ch)
```

```python
a_ch_d.to_csv('./magazine/final/a_ch_d.csv', index = False, sep ='|' , encoding = 'utf-8')
b_ch_d.to_csv('./magazine/final/b_ch_d.csv', index = False, sep ='|' , encoding = 'utf-8')
c_ch_d.to_csv('./magazine/final/c_ch_d.csv', index = False, sep ='|' , encoding = 'utf-8')
e_ch_d.to_csv('./magazine/final/e_ch_d.csv', index = False, sep ='|' , encoding = 'utf-8')
fb_ch_d.to_csv('./magazine/final/fb_ch_d.csv', index = False, sep ='|' , encoding = 'utf-8')
fn_ch_d.to_csv('./magazine/final/fn_ch_d.csv', index = False, sep ='|' , encoding = 'utf-8')
m_ch_d.to_csv('./magazine/final/m_ch_d.csv', index = False, sep ='|' , encoding = 'utf-8')
v_ch_d.to_csv('./magazine/final/v_ch_d.csv', index = False, sep ='|' , encoding = 'utf-8')
w_ch_d.to_csv('./magazine/final/w_ch_d.csv', index = False, sep ='|', encoding = 'utf-8')
```

```python
chanel_d = pd.concat([a_ch_d, b_ch_d, c_ch_d, e_ch_d, fb_ch_d, fn_ch_d, m_ch_d, v_ch_d, w_ch_d])
```
