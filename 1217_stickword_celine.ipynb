```python
# 스타일링
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'스타', '일링'", "스타일링")
# 스타일리쉬
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'스타', '일리', '쉬'", "스타일리쉬")
```

```python
# 레디 투 웨어
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'레디', '투', '웨어'", "레디투웨어")
```

```python
# 트위드
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'트', '위드'", "트위드")

# 캐주얼
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'캐, '주얼'", "캐주얼")
```

```python
# FW / SS
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'F', 'W'", "FW")
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'S', 'S'", "SS")
```

```python
# 스포티
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'스포', '티'", "스포티")

mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'벨루', '티'", "벨루티")
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'베네', '티'", "베네티")
```

```python
# 테일러 / 레이어
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'테일러', '드'", "테일러드")
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'레이어', '드'", "레이어드")

mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'에스', '닉'", "에스닉")
```

```python
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace('빈티', "빈티지")
```

```python
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'데', '일리'", "데일리")
```

```python
mag_ce['ADFN_2'] = mag_ce['ADFN_2'].str.replace("'마크', '제이콥', '스'", "마크제이콥스")
```
