### 함수 적용

```python
magazines_ce = [a_ce, b_ce, c_ce, e_ce, fb_ce, fn_ce, m_ce, v_ce, w_ce]
```

```python
def sort_by_date(df):
    df.sort_values(by='date', ascending = False, inplace = True)
    df.reset_index(drop=True, inplace = True)
```

```python
for mag in magazines_ce:
    sort_by_date(mag)
```

### check2 부여

```python
for mag in magazines_ce:
    for i in range(len(mag)):
        if not '셀린' in mag.loc[i, 'contents']:
            mag.loc[i, 'check'] = 2
```

## check 2 제외

```python
len(b_ce[b_ce['check'] != 2])
```

```python
len(b_ce[(b_ce['check'] != 2) | (b_ce['title_brand'] == 1)])
```

```python
def drop_2(df):
    return df[df['check'] != 2]
```

```python
a_ce_d = drop_2(a_ce)
b_ce_d = drop_2(b_ce)
c_ce_d = drop_2(c_ce)
e_ce_d = drop_2(e_ce)
fb_ce_d = drop_2(fb_ce)
fn_ce_d = drop_2(fn_ce)
m_ce_d = drop_2(m_ce)
v_ce_d = drop_2(v_ce)
w_ce_d = drop_2(w_ce)
```
