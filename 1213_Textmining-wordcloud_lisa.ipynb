## 트위터 패키지 안에 konlpy 모듈 호출

```python
from konlpy.tag import Twitter
```

```python
twitter = Twitter()
```

### 각 문장별로 형태소 구분하기

```python
sentences_tag = []

for sentence in article_list:
    morph = twitter.pos(sentence)
    sentences_tag.append(morph)
    
    

print(len(sentences_tag))
```

### stopword 추가

```python
sw_df = pd.read_csv('stopwords.csv', header=None)
```

```python
sw_df.columns = ['stopword']
```

```python
stop_words = []
for sw in sw_df['stopword']:
    stop_words.append(sw)
```

```python
print(len(stop_words))
print(type(stop_words))
```

```python
stop_words[-10:]
```

```python
add_0 = ['것', '수', '때', '의', '있는', '더', '그', 'I', '위', '전', 
         '이', '같은', '제', '〉', 'a', 'to', '〈', '것', '속', '↑',
        '통해', '한편', '모습', '가장', '완성', '말', '정말', '날', '곤', '노', '있어요', '중', '지금', '요', '해', '없는', '두', 
         '점', '건', '같아요', '게', '좋아하는', '듯', '최근', 'and', '오전', '입니다', '안', '후', '내', '사진']
for w in add_0:
    stop_words.append(w)
```

```python
add_1 = ['지수', '제니', '로제', 'BLACKPINK', '기자']
for w in add_1:
    stop_words.append(w)
```

```python
len(stop_words)
```

### 1. 명사, 형용사, 관형사, 알파벳, 외국어 품사만 선별해 리스트에 담기

```python
noun_adj_list= []
for sentence1 in sentences_tag:
    for word, tag in sentence1:
        if tag in ['Noun', 'Adjective', 'Alpha', 'Determiner', 'Foreign']:
            if not word in stop_words:
                noun_adj_list.append(word)
```

```python
len(noun_adj_list)
```

```python
noun_adj_list[:20]
```

```python
noun_adj_list[:10]
```

```python
ADFN_list= []
for sentence1 in sentences_tag:
    words = []
    for word, tag in sentence1:
        if tag in ['Adjective', 'Determiner','Foreign','Noun', 'Alpha']:
            if word not in stop_words:
                words.append(word)
    ADFN_list.append(words)
```

```python
len(ADFN_list)
```

```python
ADFN_list[1]
```

```python
mag_lisa['ADFN'] = ADFN_list
```

### 선별된 품사별 빈도수 계산 & 상위 빈도 10위까지 출력

```python
word = Counter(noun_adj_list)
print(word.most_common(200))
```
